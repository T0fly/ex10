In this exercise you'll try to build a neural network that predicts the price of a house according to a simple formula.

So, imagine if house pricing was as easy as a house costs 50k + 50k per bedroom, so that a 1 bedroom house costs 100k, a 2 bedroom house costs 150k etc.

How would you create a neural network that learns this relationship so that it would predict a 7 bedroom house as costing close to 400k etc.

Hint: Your network might work better if you scale the house price down. You don't have to give the answer 400...it might be better to create something that predicts the number 4, and then your answer is in the 'hundreds of thousands' etc.

```
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical

# 加载 CIFAR-10 数据集
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# 数据预处理
x_train = x_train.astype('float32') / 255.0  # 归一化
x_test = x_test.astype('float32') / 255.0    # 归一化
y_train = to_categorical(y_train, 10)       # 将标签转换为独热编码
y_test = to_categorical(y_test, 10)

# 构建 CNN 模型
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),  # 卷积层
    MaxPooling2D(pool_size=(2, 2)),                                   # 最大池化层
    Dropout(0.25),                                                   # Dropout 防止过拟合

    Conv2D(64, (3, 3), activation='relu'),                           # 卷积层
    MaxPooling2D(pool_size=(2, 2)),                                  # 最大池化层
    Dropout(0.25),                                                   # Dropout 防止过拟合

    Flatten(),                                                       # 展平
    Dense(128, activation='relu'),                                   # 全连接层
    Dropout(0.5),                                                    # Dropout 防止过拟合
    Dense(10, activation='softmax')                                  # 输出层（10 类概率分布）
])

# 编译模型
model.compile(
    optimizer=Adam(learning_rate=0.001),  # 使用 Adam 优化器
    loss='categorical_crossentropy',     # 分类交叉熵损失
    metrics=['accuracy']                 # 评估指标为准确率
)

# 训练模型
history = model.fit(
    x_train, y_train,
    batch_size=64,
    epochs=10,
    validation_split=0.2,
    verbose=1
)

# 测试模型
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"测试集损失: {test_loss:.4f}, 测试集准确率: {test_acc:.4f}")

# 保存模型
model.save("image_recognition_cnn.h5")
```

```
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# CIFAR-10 类别标签
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# 加载保存的模型
model = load_model("image_recognition_cnn.h5")

# 随机选择 10 张测试图片
indices = np.random.choice(x_test.shape[0], 10, replace=False)  # 随机选择 10 个不重复的索引

# 遍历每张图片
for i, index in enumerate(indices):
    test_image = x_test[index]                     # 获取对应图片
    true_label = np.argmax(y_test[index])          # 获取真实标签

    # 对单张图片进行预测
    test_image_expanded = np.expand_dims(test_image, axis=0)  # 增加 batch 维度
    predictions = model.predict(test_image_expanded)          # 模型预测
    predicted_label = np.argmax(predictions)                  # 获取预测的类别

    # 获取类别名称
    true_class_name = class_names[true_label]
    predicted_class_name = class_names[predicted_label]

    # 显示测试图片及预测结果
    plt.figure(figsize=(2, 2))  # 控制单张图片的大小
    plt.imshow(test_image)
    plt.title(f"True: {true_class_name}\nPredicted: {predicted_class_name}")
    plt.axis('off')
    plt.show()

    # 打印预测概率分布（可选）
    print(f"图片 {i+1}:")
    print(f"真实类别: {true_label} ({true_class_name})")
    print(f"预测类别: {predicted_label} ({predicted_class_name})")
    print(f"预测的概率分布: {predictions[0]}")
    print("-" * 50)  # 分隔线便于阅读

```
